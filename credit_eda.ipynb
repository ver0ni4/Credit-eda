```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 150)
pd.set_option('display.max_rows', 100)

%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
app = pd.read_csv('application_data.csv')
prev = pd.read_csv('previous_application.csv')

print("application_data shape:", app.shape)
print("previous_application shape:", prev.shape)
```

```python
app.info()
app.head(3)
```

```python
missing_app = app.isnull().mean().sort_values(ascending=False) * 100
missing_app = missing_app[missing_app > 0].head(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=missing_app.values, y=missing_app.index, palette='viridis')
plt.title('Топ-20 столбцов с пропусками (%) — application_data')
plt.xlabel('Процент пропусков')
plt.show()

print(missing_app)
```

```python
age_col = None
employ_col = None

for col in app.columns:
    if 'BIRTH' in col.upper() and 'DAYS' in col.upper():
        age_col = col
    if 'EMPLOY' in col.upper() and 'DAYS' in col.upper():
        employ_col = col

if age_col:
    app['AGE_YEARS'] = (-app[age_col]) // 365
    print(f"Создана колонка AGE_YEARS из {age_col}")
    app.drop(age_col, axis=1, inplace=True, errors='ignore')


if employ_col:
    app[employ_col] = app[employ_col].replace(365243, np.nan)
    app['YEARS_EMPLOYED'] = (-app[employ_col]) / 365
    print(f"Создана YEARS_EMPLOYED из {employ_col}")
    app.drop(employ_col, axis=1, inplace=True, errors='ignore')


print(app.filter(like='AGE').describe())
print(app.filter(like='EMPLOY').describe())
```

```python

threshold = 50
cols_to_drop = missing_app[missing_app > threshold].index.tolist()
app.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print(f"Удалено столбцов > {threshold}% пропусков:", len(cols_to_drop))


num_cols = app.select_dtypes(include=np.number).columns
app[num_cols] = app[num_cols].fillna(app[num_cols].median())


cat_cols = app.select_dtypes(include='object').columns
for col in cat_cols:
    app[col].fillna(app[col].mode()[0] if not app[col].mode().empty else 'Unknown', inplace=True)

print("Пропусков осталось:", app.isnull().sum().sum())
```

```python
prev_agg = prev.groupby('SK_ID_CURR').agg({
    'SK_ID_PREV': 'count',                           
    'AMT_APPLICATION': ['mean', 'max'],              
    'AMT_CREDIT': 'mean',
    'NAME_CONTRACT_STATUS': lambda x: (x == 'Approved').mean() * 100,  
    'DAYS_DECISION': 'mean'                          
}).reset_index()


prev_agg.columns = ['SK_ID_CURR', 'PREV_COUNT', 'PREV_AMT_APP_MEAN', 'PREV_AMT_APP_MAX',
                    'PREV_AMT_CREDIT_MEAN', 'PREV_APPROVED_PCT', 'PREV_DAYS_DECISION_MEAN']

print(prev_agg.head())
print("Агрегировано клиентов:", prev_agg.shape[0])
```

```python
df = app.merge(prev_agg, on='SK_ID_CURR', how='left')

df['PREV_COUNT'] = df['PREV_COUNT'].fillna(0)
df['PREV_APPROVED_PCT'] = df['PREV_APPROVED_PCT'].fillna(0)

print("Итоговая таблица shape:", df.shape)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='TARGET', data=df, palette='Set2')
plt.title('Распределение целевой переменной (TARGET)\n1 = дефолт / просрочка')
plt.xticks([0, 1], ['Нет дефолта (0)', 'Дефолт (1)'])
plt.show()

print(df['TARGET'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='AGE_YEARS', hue='TARGET', bins=20, kde=True, palette='Set1', stat='density')
plt.title('Распределение возраста клиентов по TARGET')
plt.xlabel('Возраст (лет)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='TARGET', y='AMT_INCOME_TOTAL', data=df, palette='Set3', showfliers=False)
plt.title('Доход клиентов по группам TARGET (без выбросов)')
plt.yscale('log')
plt.ylabel('Доход (лог. шкала)')
plt.show()
```

```python
corr_cols = ['TARGET', 'AGE_YEARS', 'YEARS_EMPLOYED', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 
             'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'PREV_COUNT', 'PREV_APPROVED_PCT']

corr = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Матрица корреляций ключевых признаков')
plt.show()
```

Ключевые гипотезы:

Гипотеза 1: Молодые клиенты чаще допускают дефолт  
- Да, видно по распределению возраста

Гипотеза 2: Низкий доход -> выше риск  
- Медианный доход у дефолтных ниже

Гипотеза 3: Много предыдущих заявок и низкий % одобрений -> выше риск дефолта

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='TARGET', y='PREV_APPROVED_PCT', data=df, palette='muted')
plt.title('% одобренных предыдущих заявок по TARGET')
plt.ylabel('% Approved')
plt.show()
```

Выводы:

- Почистила >50% пропусков, аномалии в стаже, преобразовала дни → годы  
- Создала удобную таблицу: объединила текущие и агрегированные предыдущие заявки  
- Главные риски: молодой возраст, низкий доход, низкий % одобрения в прошлом
